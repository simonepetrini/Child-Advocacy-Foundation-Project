```python
import requests
from bs4 import BeautifulSoup

html_content = """
<div class="xmanager_sxcomm_sx" id="container_45113"><div id="tree_menu_45118" >
<div class="treemenu_label"><h2>Commissioni permanenti</h2></div>
<ul>
<li class="OnClick"><span><a href="3447?shadow_organo=1190001" class="OnClick">1ª Affari costituzionali, affari della Presidenza del Consiglio e dell'Interno, ordinamento generale dello Stato e della Pubblica Amministrazione, editoria, digitalizzazione</a></span></li>
<li class="OnClick"><span><a href="3456?shadow_organo=1190002" class="OnClick">2ª Giustizia</a></span></li>
<li class="OnClick"><span><a href="1142?shadow_organo=1190145" class="OnClick">3ª Affari esteri e difesa</a></span></li>
<li class="OnClick"><span><a href="3698?shadow_organo=1190014" class="OnClick">4ª Politiche dell'Unione europea</a></span></li>
<li class="OnClick"><span><a href="3511?shadow_organo=1190005" class="OnClick">5ª Programmazione economica, bilancio</a></span></li>
<li class="OnClick"><span><a href="3493?shadow_organo=1190006" class="OnClick">6ª Finanze e tesoro</a></span></li>
<li class="OnClick"><span><a href="3536?shadow_organo=1190007" class="OnClick">7ª Cultura e patrimonio culturale, istruzione pubblica, ricerca scientifica, spettacolo e sport</a></span></li>
<li class="OnClick"><span><a href="3549?shadow_organo=1190146" class="OnClick">8ª Ambiente, transizione ecologica, energia, lavori pubblici, comunicazioni, innovazione tecnologica</a></span></li>
<li class="OnClick"><span><a href="3584?shadow_organo=1190147" class="OnClick">9ª Industria, commercio, turismo, agricoltura e produzione agroalimentare</a></span></li>
<li class="OnClick"><span><a href="3632?shadow_organo=1190148" class="OnClick">10ª Affari sociali, sanità, lavoro pubblico e privato, previdenza sociale</a></span></li>
</ul>
"""

# Analizza il contenuto HTML
soup = BeautifulSoup(html_content, "html.parser")

# Trova la lista delle commissioni permanenti
commissioni_list = soup.find("ul")

# Inizializza le liste per i nomi e gli URL delle commissioni
titles_list = []
hrefs_list = []

# Estrai i nomi e gli URL delle commissioni permanenti e inseriscili nelle liste
if commissioni_list:
    for li in commissioni_list.find_all("li"):
        a_tag = li.find("a")
        if a_tag:
            title = a_tag.text.strip()
            href = a_tag["href"]
            titles_list.append(title)
            hrefs_list.append(href)
else:
    print("Lista delle commissioni permanenti non trovata.")
```

```python
base_url = "https://www.senato.it/"
```

```python
hrefs_list = [base_url + href for href in hrefs_list]
hrefs_list
```

```python
titles_list = [title.split(' ', 1)[1] for title in titles_list]
titles_list
```

```python
import pandas as pd
Uffici = pd.DataFrame({'Ufficio': titles_list, 'Link': hrefs_list})
Uffici['Direzione'] = 'Senato della Repubblica'
Uffici = Uffici[['Direzione', 'Ufficio', 'Link']]
Uffici
```

```python
Uffici['Link']
```

```python
new_links = []

def get_composizione_link(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        composizione_link = soup.find("a", string="Composizione")
        if composizione_link:
            return composizione_link.get("href")
    return None

Uffici["Link_2"] = Uffici["Link"].apply(get_composizione_link)

Uffici
```

```python
Uffici["Link_2"] = base_url + Uffici["Link_2"]

Uffici
```

```python
records = []

def get_info_from_page(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        senator_divs = soup.find_all("div", class_="linkSenatore")
        for div in senator_divs:
            riferimento = div.find("a").text.strip()
            partito = div.find_all("a")[1].text.strip()
            foto_tag = soup.find("div", class_="foto").find("img")
            foto = foto_tag["src"] if foto_tag else None
            record = {
                "Link_2": url,
                "Riferimento": riferimento,
                "Partito": partito,
                "Foto": foto
            }
            records.append(record)
            
for link in Uffici["Link_2"]:
    get_info_from_page(link)
    
Riferimenti = pd.DataFrame(records)

Riferimenti
```

```python
Riferimenti["Foto"] = base_url + Riferimenti["Foto"]

Riferimenti
```

```python
Uffici_full = pd.merge(Riferimenti, Uffici, on='Link_2', how='left')
```

```python
Uffici_full = Uffici_full.drop(columns=['Link', 'Link_2'])
```

```python
nuovo_ordine = ['Direzione', 'Ufficio', 'Riferimento', 'Partito', 'Foto']
Uffici_full = Uffici_full.reindex(columns=nuovo_ordine)

Uffici_full
```

```python
Uffici_full.to_csv('senato.csv', index=False)
```
