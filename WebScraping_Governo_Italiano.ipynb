```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = "https://www.senato.it"
list_details = "/composizione/il-governo"
list_url = base_url+list_details
print(list_url)
```

```python
req = requests.get(list_url)
print(req)
```

```python
soup = BeautifulSoup(req.text)
```

```python
ministeri = soup.find_all("ul", class_="composizione")
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
lista_ruolo = []
lista_nome = []
lista_link = []

for composizione in ministeri:
    ruolo = composizione.find_previous("h3").text.strip()
    for li in composizione.find_all("li", class_=""):
        nome = li.find("a").text.strip()
        link = li.find("a")["href"]
        lista_ruolo.append(ruolo)
        lista_nome.append(nome)
        lista_link.append(link)
        

data = {'Ruolo': lista_ruolo, 'Nome': lista_nome, 'Link': lista_link}
df_ministeri = pd.DataFrame(data)

df_ministeri["Link"] = "https://www.senato.it" + df_ministeri["Link"]

pd.set_option('display.max_colwidth', None)
df_ministeri
```

```python
foto_list = []

for link in df_ministeri['Link']:
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    img_src = soup.select_one('.foto')['src'] if soup.select_one('.foto') else None
    if img_src and img_src.startswith("//"):
        img_src = "https:" + img_src
    foto_list.append(img_src)
    
df_ministeri['Foto'] = foto_list

df_ministeri
```
