```python
import pandas as pd
```

```python
camera_link = 'camera.csv'
```

```python
camera = pd.read_csv(camera_link)
```

```python
camera
```

```python
senato_link = 'senato.csv'
```

```python
senato = pd.read_csv(senato_link)
```

```python
senato
```

```python
parlamento_italiano = pd.concat([camera, senato], ignore_index=True)

parlamento_italiano
```

```python
parlamento_italiano['Istituzione'] = 'Parlamento Italiano'
```

```python
parlamento_italiano
```

```python
nuovo_ordine = ['Istituzione', 'Direzione', 'Ufficio', 'Riferimento', 'Partito', 'Foto']
parlamento_italiano = parlamento_italiano.reindex(columns=nuovo_ordine)

parlamento_italiano
```

```python
parlamento_italiano['Partito'].unique()
```

```python
for i, partito in enumerate(parlamento_italiano['Partito']):
    if partito == "FdI":
        parlamento_italiano.loc[i, 'Partito'] = "FRATELLI D'ITALIA"
    if partito == "LSP-PSd'Az":
        parlamento_italiano.loc[i, 'Partito'] = "LEGA - SALVINI PREMIER"
    if partito == "PD-IDP":
        parlamento_italiano.loc[i, 'Partito'] = "PARTITO DEMOCRATICO - ITALIA DEMOCRATICA E PROGRESSISTA"
    if partito == "M5S":
        parlamento_italiano.loc[i, 'Partito'] = "MOVIMENTO 5 STELLE"
    if partito == "Cd'I-NM (UDC-CI-NcI-IaC)-MAIE":
        parlamento_italiano.loc[i, 'Partito'] = "NOI MODERATI (NOI CON L'ITALIA, CORAGGIO ITALIA, UDC, ITALIA AL CENTRO)-MAIE"
    if partito == "Misto":
        parlamento_italiano.loc[i, 'Partito'] = "MISTO"
    if partito == "Aut (SVP-PATT, Cb)":
        parlamento_italiano.loc[i, 'Partito'] = "PER LE AUTONOMIE (SVP-PATT, CAMPOBASE)"  
    if partito == "IV-C-RE":
        parlamento_italiano.loc[i, 'Partito'] = "ITALIA VIVA-IL CENTRO-RENEW EUROPE"
    if partito == "FI-BP-PPE":
        parlamento_italiano.loc[i, 'Partito'] = "FORZA ITALIA - BERLUSCONI PRESIDENTE - PPE" 
        
parlamento_italiano
```

```python
parlamento_italiano.to_csv('parlamento_italiano.csv', index=False)
```
