```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fai una richiesta GET al sito
url = "https://lab24.ilsole24ore.com/leader-sostenibilita/2022?refresh_ce=1"
response = requests.get(url)

# Analizza l'HTML della pagina
soup = BeautifulSoup(response.text, 'html.parser')

# Trova la tabella con l'id specificato
table = soup.find('table', {'id': 'tableTmp'})

# Inizializza le liste per i dati delle colonne
aziende = []
settori = []

# Estrai i dati dalle righe della tabella
for row in table.find_all('tr'):
    columns = row.find_all('td')
    if len(columns) >= 2:
        azienda = columns[0].text.strip()
        settore = columns[1].text.strip()
        aziende.append(azienda)
        settori.append(settore)

# Crea il DataFrame
private = pd.DataFrame({'Azienda': aziende, 'Settore': settori})

private
```
