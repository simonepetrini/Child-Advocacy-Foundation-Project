```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
req = requests.get("https://www.crui.it/atenei-e-rettori-crui/elenco-rettori.html")
req
```

```python
if req.status_code == 200:
    soup = BeautifulSoup(req.text, 'html.parser')
    membri = soup.find_all("article", class_="professorView")
    
    riferimenti = []
    foto = []
    
    for membro in membri:
        riferimento_elem = membro.find("h2").find("a")
        if riferimento_elem:
            riferimento = riferimento_elem.text.strip()
            riferimenti.append(riferimento)
        else:
            riferimenti.append(None)
            
        img_elem = membro.find("img")
        if img_elem and 'src' in img_elem.attrs:
            src = img_elem['src']
            foto.append(src)
        else:
            foto.append(None)
        
    universita = pd.DataFrame({
        'Riferimento': riferimenti,
        'Foto': foto 
    })
    
else: 
    print("Errore nella richiesta HTTP:", req.status_code)
    
universita
```

```python
lista_riferimento = universita["Riferimento"].tolist()
lista_riferimento
```

```python
df = pd.DataFrame([sub.split(' - ', 1) for sub in lista_riferimento], columns=['Riferimento', 'Università'])
```

```python
df
```

```python
universita = universita.merge(df[['Riferimento', 'Università']], left_index=True, right_index=True)
universita
```

```python
universita = universita.drop(columns=['Riferimento_x'])
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
universita['Foto'] = "https://www.crui.it" + universita['Foto']
```

```python
universita = universita.rename(columns={'Riferimento_y': 'Riferimento', 'Università': 'Ufficio'})
```

```python
universita['Istituzione'] = 'Altro'
universita['Direzione'] = 'Università Italiane'
universita['Partito'] = None
```

```python
nuovo_ordine = ['Istituzione', 'Direzione', 'Ufficio', 'Riferimento', 'Partito', 'Foto']
universita = universita.reindex(columns=nuovo_ordine)

universita
```

```python
universita.to_csv('altro.csv', index=False)
```
